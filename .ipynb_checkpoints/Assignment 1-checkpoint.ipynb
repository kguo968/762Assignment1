```python
corr_matrix = house_train.corr()
corr_matrix['SalePrice'].sort_values(ascending = False)
```

Create a pairs plot of the variables with an absolute correlation value greater than 0.6. This helps to visualise (and possibly provide better insight) into how each of the variables interact with the SalePrice.

```python
df = house_train[['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea',
                  'TotalBsmtSF', '1stFlrSF', 'MSZoning']]

sns.pairplot(df, hue = 'MSZoning', plot_kws = {'alpha': 0.4})
```

As expected, there is a positive correlation between each of the included explanatory variables and the SalePrice. From the distribution plots, it is clear that majority of the observations used in the training set is from a Low density residential zone (which goes along with the bar chart plotted previously). 

Interesting to note is that there appears to be a non-linear relationship between SalePrice and OverallQual. The other explanatory variables chosen are shown to have a strongly linear relationship, although certain variables like **GrLivArea** and **1stFlrSF** seem to have a non constant variance in the distribution of the points (SalePrice variance increases as the variable value increases).

```python
_ = sns.relplot(x = 'OverallQual', y = "SalePrice", data = house_train,
           hue = 'BldgType', size = 'MSZoning',alpha = 0.5, sizes = (40,150))
```

Due to the dominance of 1FAm (single-family detatched) building types and RL (residential low density) zoned households, we aren't able to extract any useful information from the plot.

```python
facet = sns.FacetGrid(house_train, col = "MSZoning", row = "BldgType")
_ = facet.map_dataframe(sns.scatterplot, x = "OverallQual", y = "SalePrice", alpha = 0.4)
```

As seen in scatterplot above fitted with a FacetGrid, the large majority of observations recorded are in a low density residential zone, and is a single-family detached building. 

Beyond this clear standout, single-family detached buildings in a medium density residential zone, as well as townhouse end units in low and medium density residential zones are also relatively common.

Interesting to note that there appears to be an increase in SalePrice variance as OverallQual goes up for the **1Fam + RL** combinationm whereas all other combinations appear to have a mostly constant variance regardless of the OverallQual. However this may just be a result of the lack of observations present in the other combinations.

Finally it might be interesting to observe the relationship between the overall condition of the property (OverallCond), and whether the house being remodelled (YearRemodAdd) has an effect on the condition. 

In order to do this, we can split the training data set by the observations where YearRemodAdd = YearBuilt (these are the properties which haven't been remodelled) and those where YearRemodAdd $\neq$ YearBuilt, and then comparing their OverallCond values. 

```python
# Copy training data set (so that we're not working on the original data set)
house_train_remod = house_train.copy()

# Add new column indicating whether property has been remodelled or not
house_train_remod['Remod'] = np.where(house_train_remod.YearRemodAdd == house_train_remod.YearBuilt, False, True)
house_train_remod[['YearRemodAdd', 'YearBuilt', 'Remod', 'OverallCond']]
```

```python
# Plot overall condition filtered by whether it has been remodelled or not
remodFacet = sns.FacetGrid(house_train_remod, col = 'Remod')
_ = remodFacet.map_dataframe(sns.scatterplot, x = 'YearRemodAdd', y = 'OverallCond', alpha = 0.4)
```

To slightly better quantify whether there are any real differences between the two, we can instead plot on a boxplot.

```python
_ =  sns.boxplot(data = house_train_remod, x = 'Remod', y = 'OverallCond')
```

As shown from the two plots above, there does appear to be a slight difference in overall conditions between properties which have been remodelled and properties which haven't. 

While there is a single outlier in the remodelled properties, where the overall condition is 1 (the lowest in the training set), the median value is larger and the quartile range values are all larger (as expected). 

Interestingly is that because of how concentrated the OverallCond values are at 5 for the non-remodelled properties, the boxplot treats all other values as outliers, rather than plotting the upper and lower quartiles.
